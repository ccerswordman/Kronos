### 步骤 2: 指定要修改的 Python 文件路径
请将 `your_python_file.py` 替换为您实际的文件名和路径。

```python
file_path = '/content/drive/My Drive/Kronos-master/my_script.py'

# 创建一个示例文件（如果不存在），以便演示
if not os.path.exists(file_path):
    with open(file_path, 'w') as f:
        f.write("print('Hello from original script!')")
    print(f'Created a sample file at: {file_path}')
else:
    print(f'Sample file already exists at: {file_path}')

# 确认文件内容
with open(file_path, 'r') as f:
    initial_content = f.read()
print("\nInitial file content:")
print(initial_content)
```

### 步骤 3: 读取文件内容，进行修改，然后写入回文件
我们将向文件中添加一行新的打印语句。

```python
import os

# 读取当前文件内容
with open(file_path, 'r') as f:
    content = f.read()

# 添加一行新的代码
new_line = "\nprint('Hello from Colab modification!')"
modified_content = content + new_line

# 将修改后的内容写入回文件
with open(file_path, 'w') as f:
    f.write(modified_content)

print(f"File '{file_path}' modified successfully.")

# 验证修改后的文件内容
with open(file_path, 'r') as f:
    final_content = f.read()
print("\nFinal file content:")
print(final_content)
```
